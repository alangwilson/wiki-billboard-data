## wiki historical data exploration

### load in data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
fileName = "/Users/caitlynr/Documents/wiki-billboard-data/output/historical-rankings.csv"

df = pd.read_csv(fileName)
df.head()
```

### sum page views by article

```python
# overall

grouped = df.groupby('article')
pageViewsRank = grouped.sum().sort_values(by=['views'],ascending=False)
pageViewsRank.head(20)
```

```python
# split year into new column

df['year'], df['day'] = df['date'].str.split('-',1).str
df.head()
```

```python
#2016

if2016 = df['year'] == "2016"
dfOnly2016 = df[if2016]
dfOnly2016.head()
```

```python
grouped2016 = dfOnly2016.groupby('article')
pageViewsRank2016 = grouped2016.sum().sort_values(by=['views'],ascending=False)
pageViewsRank2016.head(20)
```

```python
#2017 

if2017 = df['year'] == "2017"
dfOnly2017 = df[if2017]
dfOnly2017.head()
grouped2017 = dfOnly2017.groupby('article')
pageViewsRank2017 = grouped2017.sum().sort_values(by=['views'],ascending=False)
pageViewsRank2017.head(20)
```
